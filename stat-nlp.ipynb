```
import pandas as pd
import numpy as np
```

```
path = '/content/gdrive/My Drive/imdb_reviews/preprocessed_imdbTrain.tsv'
df = pd.read_csv(path,sep='\t')
```

```
df.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.normalized_reviews,df.sentiment,test_size=0.3, random_state=42)
```

```
print('x train shape',X_train.shape)
print('x test shape', X_test.shape)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(use_idf=True, min_df=0.0, max_df=1.0)

tfidf_train = tfidf.fit_transform(X_train)

tfidf_test = tfidf.transform(X_test)
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score

multinb = MultinomialNB(alpha=1)
multinb.fit(tfidf_train, y_train)
multinb_cv_scores = cross_val_score(multinb, tfidf_train, y_train, cv=5)
multinb_cv_mean_score = np.mean(multinb_cv_scores)

print('Mean CV Accuracy:', multinb_cv_mean_score)
multinb_test_score = multinb.score(tfidf_test, y_test)
print('Test Accuracy:', multinb_test_score)
```

```
from sklearn.metrics import confusion_matrix

y_pred = multinb.predict(tfidf_test)
confusion_matrix(y_test, y_pred)
```
